```python
import numpy as np
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.read_csv('solar_power.csv')
df
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
model = RandomForestClassifier(n_estimators=100)
model.fit(Xtrain,ytrain)
```

```python
print('Training efficieny:',model.score(Xtrain,ytrain))
print('Testing efficieny:',model.score(Xtest,ytest))
```

```python
with open('model.pkl','wb') as file:
    pickle.dump(model,file)
```
